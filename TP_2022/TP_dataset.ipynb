```python
import numpy as np
import csv
f = open('5호선_2019.csv', 'r' ,encoding="utf-8")
data= csv.reader(f, delimiter=',')
print(data)
```

```python
data_dict=dict()
head = list()
```

```python
num = 0

for row in data:
    num += 1
    
    if num == 1:
        head = row
        continue
    
    if row[0] not in data_dict.keys():
        data_dict[row[0]] = list()
        
        list_row = row
        for j in range(4, 24):
            list_row[j] = int(list_row[j])
        
        data_dict[row[0]].append(list_row)
        continue
    
    else:
        n2 = 0
        
        for idx in range(len(data_dict[row[0]])):
            if data_dict[row[0]][idx][1] == row[1] and data_dict[row[0]][idx][2] == row[2] and data_dict[row[0]][idx][3] == row[3]:
                n2 = 1
                for j in range(4, 24):
                    data_dict[row[0]][idx][j] += int(row[j])
                break
                
        if n2 == 0:
            list_row = row
            for j in range(4, 24):
                list_row[j] = int(list_row[j])
            data_dict[row[0]].append(list_row)
            
print("complete")
    
```
