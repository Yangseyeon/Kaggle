```python
train_data = pd.read_csv('/kaggle/input/2022-ai-w5p1/train.csv')
test_data = pd.read_csv('/kaggle/input/2022-ai-w5p1/test.csv')
sample_data = pd.read_csv('/kaggle/input/2022-ai-w5p1/sample_submit.csv')
```

```python
train_y = train_data['label']
train_x = train_data.drop('label', axis = 1)
train_x
```

```python
import torch
import torch.optim as optim

torch.manual_seed(1)
x_train = torch.FloatTensor(train_x.values)
x_test = torch.FloatTensor(test_data.values)
y_train = torch.LongTensor(train_y.values)


y_train
```

```python
x_train
```

```python
nb_class = 10
nb_data = len(y_train)

y_one_hot = torch.zeros(nb_data, nb_class)
y_one_hot.scatter_(1, y_train.unsqueeze(1), 1)

print(y_one_hot)
```

```python
import torch.nn.functional as F  # for softmax

nb_class = 10
nb_data = len(y_train)

W = torch.zeros((784, nb_class), requires_grad=True)
b = torch.zeros(1, requires_grad=True)
# optimizer 설정
optimizer = optim.SGD([W, b], lr=0.00001)
nb_epochs = 2000
for epoch in range(nb_epochs + 1):

    # Cost 계산 (1)
    hypothesis = F.softmax(x_train.matmul(W) + b, dim=1) # or .mm or @
    
#     print(hypothesis.shape)
#     print(y_one_hot.shape)
    

#     cost = (y_one_hot * -torch.log(hypothesis)).sum(dim=1).mean()
    
    # cost 표현법 2번 예시
    # cross_entropy를 사용하면 scatter 함수를 이용한 one_hot_encoding을 안해도 됨
    cost = F.cross_entropy((x_train.matmul(W) + b), y_train)

    # cost로 H(x) 개선
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    # 100번마다 로그 출력
    if epoch % 200 == 0:
        print('Epoch {:4d}/{} Cost: {:.6f}'.format(
            epoch, nb_epochs, cost.item()
        ))
```

```python
hypothesis = F.softmax(x_train.matmul(W) + b, dim=1)
predict = torch.argmax(hypothesis, dim=1)
print(hypothesis)
print(predict)
```

```python
hypothesis = F.softmax(x_test.matmul(W) + b, dim=1)
predict = torch.argmax(hypothesis, dim=1)
print(hypothesis)
print(predict)
```
