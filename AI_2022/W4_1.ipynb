```python
train_data = pd.read_csv('/kaggle/input/2022-ai-w4p1/train.csv')
test_data = pd.read_csv('/kaggle/input/2022-ai-w4p1/test.csv')
```

```python
train_y = train_data['Diabetes']
train_x = train_data.drop('Diabetes', axis = 1)
train_x
```

```python
import torch
import torch.optim as optim

torch.manual_seed(1)
x_train = torch.FloatTensor(train_x.values)
x_test = torch.FloatTensor(test_data.values)
y_train = torch.FloatTensor(train_y.values)


y_train = y_train.view([-1, 1])

y_train.shape
```

```python
x_train.shape
```

```python
# 모델 초기화
W = torch.zeros((8, 1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)

# optimizer 설정
optimizer = optim.SGD([W, b], lr=0.0001)

nb_epochs = 500000

loss = torch.nn.BCELoss()

for epoch in range(nb_epochs + 1):

    # Cost 계산
    hypothesis = torch.sigmoid(x_train.matmul(W) + b) # or .mm or @
    
#     print(hypothesis.shape)
    
    cost = loss(hypothesis, y_train)
    # cost로 H(x) 개선
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    # 100번마다 로그 출력
    if epoch % 10000 == 0:
        print('Epoch {:4d}/{} Cost: {:.6f}'.format(
            epoch, nb_epochs, cost.item()
        ))
```

```python
hypothesis = torch.sigmoid(x_test.matmul(W) + b)
prediction = hypothesis >= torch.FloatTensor([0.5])
# prediction
```

```python
submit = pd.DataFrame()

submit['ID'] = pd.Series(range(len(prediction)))
submit['Diabetes'] = prediction.detach().numpy()

submit
# submit.to_csv('submit.csv', index = None)
```
